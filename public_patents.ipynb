analyse json data pulled from google public patents

```python
import json
import pandas as pd
```

For simplicity, code is duplicated to see results from the two different searches: by keyword and by assignee

## Keyword

```python
json_input = "json_data/keyword_search_data_20211126.json"
# json_input = "json_data/assignee_search_data_20211203.json"

all_data = []
for line in open(json_input, 'r'):
    all_data.append(json.loads(line))
```

```python
df = pd.DataFrame(all_data)
```
